### Variables en lista

```python
##Lista en Pythonn:Estructura de datos mutable
vendedores = ["Juan", "Ana", "Pedro", "Luisa"]
print(vendedores)
print(type(vendedores))

##Modificar elemento de la lista
vendedores[1]="Sebastian"
vendedores[1]+=" Panda"
vendedores.append("Duber")
print(vendedores)

##Tupla en Python: Estrctura de datos inmutable
vendedores1 = ("Juan", "Ana", "Pedro", "Luisa")
print(vendedores1)
print(type(vendedores1))

##Listas vacias
meses= []
ventas=[]

cantidad_meses=int(input("Ingrese el numero de meses a registrar:"))

for i in range (cantidad_meses):

    mes=input("Ingrese el nombre del mes:")
    venta=float(input(f"Ingrese el total de ventas en {mes}: "))

    ## Guardar los datos
    meses.append(mes)
    ventas.append(venta)

n=len (meses)

print("\nRegistro de Ventas mensuales:")
for i in range (n):
    print(f"En el mes de {meses[i]} se registraron ventas por un total de: ${ventas[i]:.2f}")
```

```python
##Uso de listas para almacenar nombres y ventas de vendedores
vendedores = ["Juan", "Ana", "Pedro", "Luisa"]
ventas = [1500.50, 2300.75, 1800.00, 2100.25]

for i in range(len(vendedores)):
    print(f"El vendedor {vendedores[i]} tuvo un total de ventas de: ${ventas[i]:.2f}")
```

```python
###Uso de pandas para manejo de datos
import pandas as pd
vendedores = ["Juan", "Ana", "Pedro", "Luisa"]
ventas = [1500.50, 2300.75, 1800.00, 2100.25]
df=pd.DataFrame({
    "Vendedor":vendedores,
    "Ventas":ventas
})
print(df)
```
